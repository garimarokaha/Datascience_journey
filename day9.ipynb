```python
=> simple linear regression: there is only one independent variable.
   y = b0 + b1x

=> multiple linear regression : more than one independent variable
    y = b0 + b1x1 + b2x2
```

```python
data = pd.read_csv('Salary Data.csv')
data.head()
```

```python
#data bata anull values hataunaa 
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
x = data[['Age']]
y = data.Salary
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=6)
```

```python
model = LinearRegression()
model.fit(xtrain,ytrain)

```

```python
model.intercept_
```

```python
model.coef_
```

```python
yprediction = model.predict(xtrain)  #train, test jun dida ni vayo
```

```python
plt.plot
```

```python
plt.scatter(data.Age,data.Salary)
plt.plot(xtrain,yprediction , color = 'red') #line aauna ko lagi bich ma regresson ko
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
yprediction_test = model.predict(xtest)
```

```python
plt.scatter(xtest,ytest)
plt.plot(xtest,yprediction_test , color = 'red') #line aauna ko lagi bich ma regresson ko

#plot garnaa ko lagi jun data legera apredict garechham tehi lenu oparchhaa, suppose train legera
#predict gareko vaye xtrain garnu paryo , xtest leko vaye xtest garnu paryo plot maaa 
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

### multiple linear regression

```python
x = data[['Age','Years of Experience']]
y = data.Salary
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2 , random_state =9)
```

```python
xtrain
```

```python
xtest
```

```python
ml_model = LinearRegression()
ml_model.fit(xtrain,ytrain)
```

```python
yehaa baneko eqn bhaneko: b0 + b1*age + b2 * years of experience
```

```python
ml_model.intercept_   # yo bhaneko b0 ko value aako ho
```

```python
ml_model.coef_ # b1 ra b2 ko value
```

```python
#age = 30
#years of expereience = 5

pred_salary = ml_model.predict([[30,5]])
print(pred_salary)
```

```python
#Evaluation
yprediction = ml_model.predict(xtest)
```

```python
yprediction
```

```python
#Mean absolute Error
mae = mean_absolute_error(ytest,yprediction)
print('MAE:',mae)
```

```python
# R _squared error
R2 =r2_score(ytest,yprediction)
print('R2:',R2)
```

```python
multiple linear regression ko assumption 
1) independent ra dependent variable ko bicha ma linear relationship hunu parchhaa 
   independent _dependent variable : linear relationship
    (age-Salary)
     (yoe,salary)

2) No multicollinearity : independent variable ko yek arkaaa snagaa corelation small hunu parchhaa(relation nahuney independent variable bichhaa)
   -> independent variables (age,yoe) no correlation 
```

```python
#2>collinearity herna ko lagi
corr = data['Age'].corr(data['Years of Experience'])
print(corr)
```

```python
#correlation high aayo yehaa yesko arthaa 2 otaa variable chaidenaa. yeutaa pass gardei hunchhaa vannaa khojeko
```
