```python
#regression Evaluation metrices

1. MSE (Mean Squared Error)
   -> aayeko sabbei error ko square garenchha aaba chai tesko mean nekalaenchha
   -> first square of mean
   -> second ma chai tesko error nekalencha

2.  RMSE(Root mean square error) 
  -> MSE lai square root lagaidenchhaa
  -> big range of error lai small ma lyaunaa ko lagi help garchhaa

3. Mean absolute error
  -> absolute value of error 
-> negative error lai positive garauchhaa


4. MAPE

5. R2
```
