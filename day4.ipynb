```python
missing value ko lagi 2 otaa option hunchhaa
1) fill
2) Drop_

If any column , most of the value are missing then it doesn't have any significant in our data then remove columnn 

dherei data missing vaye => delete that column
thorei matraaa value missing vaye 1) remove that row
                                  2) fill the data (fill garda {mean (),median(),mode(),forwardfill(),backwardfill(),nearestvalue}
                                                    ,value ley fill garney)
```

```python
forward fill()=> aagadi ko value ley fill gardeney
eg: 172 ma value chhaina bhane 171 ma bhako value ley replace gardeney


backward fill()=> pachhadi  ko value ley fill gardeney
eg: 172 ma value chhaina bhane 173 ma bhako value ley replace gardeney


Nearest value =>arko kunei row khojenchhaa jasko value similar chhaa
```
