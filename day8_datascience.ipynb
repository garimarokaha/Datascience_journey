```python
portion of variance of depenedent variable , explained by independent variable => r2 square 
independent variable ley dependent variable lai kati explain garnaa sakerachhaa
```

```python
#syntax: mean_square-error(yactual,ypredicated)
```

```python
y actual : y ma chhaa
y predicted : ypred ma chhaa
```

```python
#mean swuare error
mse = mean_squared_error(y,ypred)
print(f'MSE:{mse:.2f}')
```

```python
RMSE ko lagi direct function chhaina so 
mse ko square root  garnaa saknecha
```

```python
#root mean square error
rmse = np.sqrt(mse)
print(f'RMSE:{rmse:.2f}')
```

```python
#mean absolute error
mae = mean_absolute_error(y,ypred)
print(f'MAE:{mae:.2f}')
```

```python
#mean absolute percentage error
mape = mean_absolute_percentage_error(y,ypred)
mape=mape*100
print(f'MAPE:{mape:.2f}%')
```

```python
#R2 square
r2 = r2_score(y,ypred)
print(f'R2_score:{r2:.2f}')
```
