```python
# steps in the text mining 
1. Tokenization=>text ko yauta word chai yauta token jasto banauno lai tokenization
2. Most Frequent=>we can also see most frequent words and other  from the text given document 

3. puncutation removmal=> text bhayak ko sabai lai remove garidinxa(likewise ,.'''')

4. stopwords removal=>conjuction and preposition(a, an, the , or ,but)

5.stemming/lemmatization: 
    stemming: prefix haru lai hatayedinxa 
    Lemmatization:  back to root word banaidinxa(jastai, run, ran ,running xa teslai chai root word run banaidixa)
    
6. pos tagging=> parts of speech tagging 
-POS tagging is the process of assigning a part of speech (e.g., noun, verb, adjective) to each
word in a sentence based on its context and definition.
```
