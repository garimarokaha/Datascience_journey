```python
#Applying logistic regression on this 'iris ' datasets
x = data[['sepal_length','sepal_width','petal_length','petal_width']]
y = data.species
```

```python
# or yesto garnaa sakenchhcaa yeutaa species lai drop gardeney : baki raheko x
x = data.drop('species', axis=1)
x
#column ko lagi axis = 1
# row ko lagi axis =0
```

```python
x = [['sepal_length', 'sepal_width', 'petal_length', 'petal_width',
       'species']]   #yeasri gardda pani vayo
```

```python
#train ,test split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2,random_state=16)
```

```python
#model fitting
model = LogisticRegression(multi_class ='multinomial') #multi_claass hudaa yesari parameter fit garne vanchhaa
model.fit(xtrain,ytrain)
```

```python
#sepal_length = 6
#sepal_length = 3
#petal_length = 5
#petal-width = 1

model.predict([[6,3,5,1]])
```

```python
help(LogisticRegression)
```

```python
yprediction = model.predict(xtest)
yprediction
```

```python
#Accuracy
accuracy = accuracy_score(ytest,yprediction)
print(f'Accuracy:{accuracy:.2%}')
```

```python
#confusion matrix
cm = confusion_matrix(ytest,yprediction)
cm
```

```python
confusion matrix shape = n * n , where n is number of class label
```

```python
#better visusalization ko lagi
sns.heatmap(cm, annot = True , cbar = False , cmap ='viridis')  #annot = true ley gardda tyo vitraa number aako
plt.show()
```
