```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix , accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
```

```python
data = sns.load_dataset('iris')
```

```python
sns.get_dataset_names()
```

```python
data.head()
```

```python
now yo data ma k nearest neighbor lagaidei chham
```

```python
x = data.drop('species',axis =1)   #species bahek ko aru sabei lai x banaunaa ko lagi
y = data.species                    #species lai y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2,random_state=4)
```

```python
knn= KNeighborsClassifier(n_neighbors = 9)  # n_neighbors bahneko k ko value deney  
 #default = 5 lenchhaa.. ra jahely pani odd denu parchhaa 
knn.fit(xtrain,ytrain)
```

```python
help(KNeighborsClassifier)
```

```python
ypred = knn.predict(xtest)
ypred
```

```python
cm = confusion_matrix(ytest,ypred)
```

```python
cm
```

```python
accuracy = accuracy_score(ytest,ypred)
print(f'Accuracy:{accuracy*100}')
```
