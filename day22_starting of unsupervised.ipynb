```python

from sklearn.cluster import KMeans
```

```python
#fit garnee aaba
Kmeans = KMeans(n_clusters=5)  
Kmeans.fit(x)                 
```

```python
#indivual dataa kun clusters ma parechhaa vaneraa vndenchhaa. 
label=Kmeans.labels_
label
```

```python
plt.scatter(x.annual_income , x.spend_score , c= label) # c - label gareko harek cluster lai different coloir ko lagi
plt.show()
```
