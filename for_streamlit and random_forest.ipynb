```python
import pandas as pd
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('Diabetes.csv')
data.head()
```

```python
x = data[['Glucose','BloodPressure', 'DiabetesPedigreeFunction','Age']]
```

```python
x
```

```python
y = data.Outcome
```

```python
y
```

```python
xtrain, xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2 , random_state= 90)
```

```python
dmodel = DecisionTreeClassifier(max_depth = 7)
dmodel.fit(xtrain, ytrain)
```

```python
#model evaluation
yprediction = dmodel.predict(xtest)
```

```python
accuracy = accuracy_score(ytest,yprediction)
print(f'Accuracy = {accuracy:.2%}')
```

```python
#Glucose = 200
# Blood pressure = 140
#DFP = 0.9
#Age = 40

dmodel.predict([[200,140,0.9,40]])
```

```python
Decision tree ko first sampling ma , jasta ko tastei 10 ottaa row ledenaa.
bhannu ko meaning yeutei 2 patak repaet hunaa ni payo j garnaa ni payo randomly 10 ottaaa pick garchhaa
, kunei rows multiple time sleect hunaa ni sakchhaa, kunei select nai nahuna pani saskchhaaa. 



ie: Randomsampling -- select 10 rows 
```

```python
Boot Strap Samples --> total maddhe yeutaaa lai randomly select garneee ani oick garneee, pick vaisakkeo mannchee pani feri basney line ma

eg: 6 otaa choclate badhha, sir ley first ma malai sekect garnu vayo , aarko patak feri ma bahira gayeraa basne hainaa tehi basneyy, arko hhoti badhhaa 
ma parnaa ki sakchhu na parnaa ni sakchhuu.
```
