```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error , mean_absolute_percentage_error, r2_score
from sklearn.model_selection import train_test_split
from mpl_toolkits.mplot3d import Axes3D
```

```python
df =pd.read_csv('Salary Data.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python


x = df[['Age', 'Years of Experience']]
x.head()
```

```python
y = df[['Salary']]
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=7)
```

```python
#fitting the model
linear_model = LinearRegression()
linear_model.fit(x,y)
```

```python
linear_model.intercept_
```

```python
linear_model.coef_
```

```python
predictions =linear_model.predict(x)
print(predictions)
```

```python
#test ko matraa data ko lagi
ypred = linear_model.predict(xtest)
ypred
```

```python
#error nekalnee
# mean_absolute_error
mae = mean_absolute_error(ytest,ypred)
print(f'Mean_absolute_eror:{mae:.2f}')
```

```python
#root man squared error (RMSE)
rmse = np.sqrt(mae)
print(f'Root_Mean_squared_eror:{rmse:.2f}')
```

```python
#R _ squared (coefficient of determination)
r2 = r2_score(ytest,ypred)
print(f'r2_score:{r2:.2f}')
```

```python
print("True values:",ytest.values)
print("predicted values:",ypred)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming x is your DataFrame with 'Age' and 'Years of Experience'
# and model is your trained linear regression model

# Create a meshgrid for prediction
age_range = np.linspace(x['Age'].min(), x['Age'].max(), 50)
exp_range = np.linspace(x['Years of Experience'].min(), x['Years of Experience'].max(), 50)
Age_mesh, Exp_mesh = np.meshgrid(age_range, exp_range)
grid_points = np.c_[Age_mesh.ravel(), Exp_mesh.ravel()]
salary_pred_plane = linear_model.predict(grid_points).reshape(Age_mesh.shape)

# Regression line along Age = Years of Experience
line_vals = np.linspace(max(x['Age'].min(), x['Years of Experience'].min()), 
                        min(x['Age'].max(), x['Years of Experience'].max()), 100)
line_features = np.column_stack((line_vals, line_vals))  # Age = Years of Experience
salary_pred_line = linear_model.predict(line_features)  # Use the same model consistently

# Plotting
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of actual test data
ax.scatter(xtest['Age'], xtest['Years of Experience'], ytest, color='blue', label='Actual Salary')

# Plot regression plane
ax.plot_surface(Age_mesh, Exp_mesh, salary_pred_plane, color='red', alpha=0.4)

# Plot regression line along Age=Experience
ax.plot(line_vals, line_vals, salary_pred_line, color='black', linewidth=3, label='Regression Line (Age=Experience)')

# Labels and title
ax.set_xlabel('Age')
ax.set_ylabel('Years of Experience')
ax.set_zlabel('Salary')
ax.set_title('Multiple Linear Regression: Plane + Regression Line')
ax.legend()

plt.show()
```
