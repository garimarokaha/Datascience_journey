```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.tree import DecisionTreeClassifier ,plot_tree
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
```

```python
data = sns.load_dataset('iris')
data.head()
```

```python
x = data.drop('species',axis =1)
y = data.species
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2, random_state =7)
```

```python
dmodel = DecisionTreeClassifier()
dmodel.fit(xtrain,ytrain)
```

```python
#testing ko accurcy nekalne
```

```python
yprediction = dmodel.predict(xtest)
accuracy = accuracy_score(ytest,yprediction)

print(f'Accuracy:{accuracy:.2%}')
```

```python
#training accuracy
ypred_train = dmodel.predict(xtrain)
accuracy_train = accuracy_score(ytrain,ypred_train)

print(f'Accuracy:{accuracy_train:.2%}')
```

```python
train ma sabeidata meleko chhaa so acuracy 100%, yo pani suspecious ho so overfitting.
```

```python
features = list(x.columns)
clas = list(y.unique())
```

```python
plot_tree(dmodel,feature_names=features,class_names=clas,filled = True)
plt.show()
```
