### Data Scaling

```python
Data Scaling

-> scaling bhaneko k ho vandaa , hami snagaa vako column of variable ko
data ko range yeutei chhaina chhutaa chhutei chha bhaney we should use data scaling

-> suppose hami sangaa 2 ottaa columns cha ie; 
     1) AGE (range : 21-80)
     2) Insulin (Insulin : 0-450)
yo 2 ottaa bataa hami lai diabetes predict garnu chha,

-> yeni haru range nai farak farak chha
-> suppose we have 1 data: Age = 40, insulin = 300
   suppose we have another  data: Age = 42, insulin = 100 chha
  now KNN ley  ta diatance nekalertaa garchhaa,
     distance = root over (40-42)^2 +(300-100)^2
              =root over 2^2 + (-200)^2
yeutaa ma 2 ko square 4
arko ma 200 ko square gardaa 2000000

-> yesto value aauda high value aako ley chhai sano value lai dominant garchha
-> diatance based ko calculation chaa bahney yeni ahru lai range ma lyaunu parchha
ani ballaa distance formlulaa lagaune ; ie: same rangee ma lyaunee 
this process is called data scaling . 
```

### Data scaling
-> data set haru lai commom range ko banaune.

Data scaling techniques

1) Min _ Max Scaling (Normalization)
   -> values lai convert gareraa 0 to 1 ko bich ma aauchhaa 
   -> formula : (x-min)/(max-min)
   
3) Standardization (Z - score Normalization)
   -> mean = 0 and std = 1

   -> formula : x -meu / sigma

   ie: z = score - mean / SD(standard deviation)

   -> normally bahnney vane -1 to 1 ko range ma aauchha. 


```python
#Data scaling

1.MinMax Scaling
   xscaled =(x - xmin) / (xmax -xmin) : range(0-1)
-> yeslai aru range mani convert garna parnaa sakchha ie jasto hamalai 2-10 ko range ma chaiyo bahney
    fromula :xnew_min+((x-xmin)/(xmax-xmin)) * (xnew_max-xnew_min)
        here xnew min = 2
             xnew max =10

2. Z - score normalization
  scaled = x -min/std
```
